```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.array([[1.0, 1.0, 'OR'],[1.0, 2.0, 'OR'],[1.0, 2.5, 'OR'],[3.0, 3.0, 'GF'],[3.5, 2.5, 'GF'],[1.5, 1.5, 'OR'],[2.0, 2.0, 'OR'],[2.5, 2.5, 'OR'],[3.0, 4.0, 'GF'],[3.5, 3.0, 'GF'],[4.5, 3.5, 'GF'],[1.2, 1.2, 'OR'],
    [1.8, 1.8, 'OR'],[2.2, 2.3, 'OR'],[3.3, 3.3, 'GF'],[3.7, 2.8, 'GF'],[4.1, 3.1, 'GF'],[1.1, 1.3, 'OR'],[1.3, 2.1, 'OR'],[1.9, 2.6, 'OR'],[3.2, 3.1, 'GF'],[3.6, 2.9, 'GF'],[4.2, 3.2, 'GF'],[1.4, 1.4, 'OR'],
    [2.1, 2.1, 'OR'],[2.6, 2.6, 'OR'],[3.1, 4.1, 'GF'],[3.4, 3.2, 'GF'],[4.3, 3.6, 'GF'],[3.,2.5,'NN']])
query = [3.0,2.5,'NN']
```

```python
df = pd.DataFrame(data)
df.columns = ['x','y','cat']
df
```

```python
for i in range(7):
    if(df.iloc[i]['cat'] == 'GF'):
        plt.scatter(df.iloc[i]['x'], df.iloc[i]['y'], s=100, c='b')
    elif(df.iloc[i]['cat'] == 'OR'):
        plt.scatter(df.iloc[i]['x'], df.iloc[i]['y'], s=100, c='r')
    else:
        plt.scatter(df.iloc[i]['x'], df.iloc[i]['y'], s=200, c='y')

plt.grid()
plt.show()
```

```python
import math

dis = []
for i in range(30):
    if pd.isna(df.iloc[i]['x']) or pd.isna(df.iloc[i]['y']):
        continue
    x_diff = float(df.iloc[i]['x']) - query[0]
    y_diff = float(df.iloc[i]['y']) - query[1]
    dis.append(math.sqrt(x_diff**2 + y_diff**2))
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```
